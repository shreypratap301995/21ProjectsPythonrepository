```python
heartdisease_df['trestbps'].fillna(heartdisease_df['trestbps'].mean(), inplace=True)
print("Missing values after filling 'trestbps':")
print(heartdisease_df.isnull().sum())
```

the codeabove fills  the missing values in the trestbps column of the heartdisease_df DataFrame with the mean of the existing values in that column. The inplace=True argument modifies the DataFrame directly. It also prints the count of missing values for each column after filling trestbps.
